# Data Cleaning:

1. Analyse the data and identify which columns are not relevant for survivor prediction task. Drop those 
columns from the dataframes. 

```python
training_df.drop('Ticket',inplace = True, axis = 1)
training_df.drop('PassengerId',inplace = True, axis = 1)
training_df.drop('Cabin',inplace = True, axis = 1)
training_df
```

Ticket, PassengerId and Cabin Columns are not relevant for survivor prediction task

2. Check how many columns have missing values in them (NA) and how many have NaN values. Logically 
impute the dataset.

```python
training_df.isna().sum()
```

```python
training_df['Age'] = training_df['Age'].fillna(0)
training_df['Embarked'] = training_df['Embarked'].fillna(0)
training_df
```

3. Identify any categorical valued columns (non-numeric) and convert them to numeric. 

Converting Sex Column from male = 0 and female = 1;

```python
training_df.loc[training_df["Sex"] == "male", "Sex"] = 0
training_df.loc[training_df["Sex"] == "female", "Sex"] = 1
training_df
```

Converting Embarked Column from S = 0, C = 1 and Q = 2;

```python
training_df.loc[training_df["Embarked"] == "S", "Embarked"] = 0
training_df.loc[training_df["Embarked"] == "C", "Embarked"] = 1
training_df.loc[training_df["Embarked"] == "Q", "Embarked"] = 2
training_df
```
